```python
#2-

l=[['alice','99300063'],['Bob','53124596']]

 
def nouveau(nom,numero):
    """ pour entrer un nouveau contact dans le carnet d'adresse"""
    if est_present(nom):
        print("Nom déjà présent")
    else:
        l.append([nom, numero])
     
def est_present(nom):
    """ pour vérifier si le nom n'est pas déjà existant"""
    for liste in l:
        name, tel = liste
        if name == nom:
            return 1
    return 0
 
def afficher():
    #pour afficher la liste de tous les noms et numéros.
    for liste in l:
        name, tel = liste
        print("{0} -> {1}".format(name, tel))
     
     
def cherche_nom(nom):
    """ pour chercher le numéro avec le nom"""
    for liste in l:
        name, tel = liste
        if name == nom:
            return tel
    
 
    
     
 
def carnet():
    print("1 : entrer un nouveau contact \n 2 : chercher un numéro")
    print("3 : afficher la liste \n 4: quitter")
   
 
    choix=input("votre choix : ")
    if choix =="1":
        nom=input("nouveau contact :")
        num=input("numéro de téléphone :")
        nouveau(nom,num)
        carnet()
    elif choix=="2":
       
        recherche=input("nom de la personne ?")
        print(cherche_nom(recherche))
        carnet()
    elif choix=="3":
        afficher()
        carnet()
    elif choix=="4":
        print('au revoir ...')
    else :
        carnet()
 

         
carnet()
```
